## **Detecting and Handling outliers**

#### **`What are outliers?`**
Outliers are data points that are significantly different from the rest of the data in a dataset. They can be much higher or much lower than most values and can affect analysis, such as averages, correlations, or model predictions.
#### **`How to detect outliers?`**
We can detect outliers in using:
- Boxplot ‚Äì Visualize data; points outside the whiskers are outliers.
- IQR Method ‚Äì Values below Q1 - 1.5*IQR or above Q3 + 1.5*IQR are outliers.
- Z-Score Method ‚Äì Values with |Z| > 3 are considered outliers.
#### **`How to handle outliers?`**
We can handle outliers in these ways:
- Remove them ‚Äì Delete if they are errors or irrelevant.
- Cap or floor them ‚Äì Replace extreme values with a maximum or minimum threshold.
- Transform data ‚Äì Apply log, square root, or other transformations to reduce their impact.
- Keep them ‚Äì If they are valid observations, sometimes it‚Äôs better to leave them.

`Here, we use a boxplot to detect outliers and apply log transformation and capping to handle them.`

**`Still outliers are present in columns which are Rented Bike count, Rainfall, Snowfall, Wind Speed.`**

```python
# Now we are using capping method to handle remaining outliers
# Capping is a way to handle outliers by replacing extreme values with a maximum or minimum limit to reduce their effect on analysis.

columns = ['Rented Bike Count', 'Wind speed (m/s)', 'Solar Radiation (MJ/m2)', 'Rainfall(mm)', 'Snowfall (cm)']
for col in columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = np.clip(df[col], lower, upper)
```

## **Feature Scaling**

##### **`What is feature scaling?`**
Feature scaling means resizing numerical data so all features are on a similar scale. This helps machine learning models work better and faster.
##### **`Methods for Feature Scaling`**
- Min-Max Scaling ‚Äì Scales values to a fixed range, usually 0 to 1.
- Standardization (Z-score Scaling) ‚Äì Scales data to have mean 0 and standard deviation 1.
- Max Abs Scaling ‚Äì Scales data by dividing by the maximum absolute value, keeping values between -1 and 1.
- Robust Scaling ‚Äì Uses median and IQR, reducing the effect of outliers.

```python
# Here we are using Robust Scaling because in this datset their are lots of outliers present

from sklearn.preprocessing import RobustScaler   # importing robust scaler from skleran.preprocessing
scaler = RobustScaler()                     # initializing robust scaler

# List of columns with numerical values
cols = ['Temperature(¬∞C)', 'Humidity(%)',
       'Wind speed (m/s)', 'Visibility (10m)', 'Dew point temperature(¬∞C)',
       'Solar Radiation (MJ/m2)', 'Rainfall(mm)', 'Snowfall (cm)']

# Applying robust scaler to each numerical columns
df[cols] = scaler.fit_transform(df[cols])
```
